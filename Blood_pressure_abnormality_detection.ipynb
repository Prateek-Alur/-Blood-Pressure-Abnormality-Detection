# Feature selection

```python
importances = rfc.feature_importances_
feature_names = train_bp_x.columns
feature_importances = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

print("\nFeature Importances:")
print(feature_importances)
```

```python
top_n = 9  
selected_features = feature_importances['Feature'][:top_n].tolist()
selected_features
```

```python
selected_features = ['Genetic_Pedigree_Coefficient',
 'Level_of_Hemoglobin',
 'Chronic_kidney_disease',
 'Adrenal_and_thyroid_disorders',
 'salt_content_in_the_diet',
 'Physical_activity',
 'Age',
 'alcohol_consumption_per_day',
 'BMI']


train_bp_x_selected = train_bp[selected_features]  
test_bp_x_selected = test_bp[selected_features]


train_bp_y = train_bp.iloc[:, -1]  # Last column is the target
test_bp_y = test_bp.iloc[:, -1]
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform it
train_bp_x_scaled = scaler.fit_transform(train_bp_x_selected)

# Transform the test data using the same scaler
test_bp_x_scaled = scaler.transform(test_bp_x_selected)
```

```python
from sklearn.linear_model import LogisticRegression
logreg_bp= LogisticRegression()
logreg_bp.fit(train_bp_x_scaled, train_bp_y)
```

```python
pred_test_bp= logreg_bp.predict(test_bp_x_scaled)
```

```python
from sklearn.metrics import confusion_matrix
tab1= confusion_matrix(test_bp_y, pred_test_bp)
tab1
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_bp_y, pred_test_bp))
print('Accuracy is --->', tab1.diagonal().sum()*100/tab1.sum())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import GridSearchCV

# Define the RandomForestClassifier
rfc = RandomForestClassifier(random_state=42)

# Define the hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 500],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(train_bp_x_scaled, train_bp_y)

# Best parameters and model
best_rfc = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)

# Fit the best model
best_rfc.fit(train_bp_x_scaled, train_bp_y)

# Predictions and evaluation
pred_test = best_rfc.predict(test_bp_x_scaled)
tab1 = confusion_matrix(test_bp_y, pred_test)
print(tab1)
print(classification_report(test_bp_y, pred_test))
print('Accuracy is --->', tab1.diagonal().sum() * 100 / tab1.sum())
```
